```python

import numpy as np
from matplotlib import pyplot as plt
x = np.arange(1,11)
y=2*x+5
plt.title("Line Plot")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.plot(x,y)
plt.show()
plt.plot(x,y,"or")
plt.grid()
plt.grid(color = 'red',linestyle = '--',linewidth=0.8)
plt.show()
```

```python

import numpy as np
from matplotlib import pyplot as plt
x = np.arange(0,3*np.pi,0.1)
y=np.sin(x)
plt.title("sin wave plot")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.plot(x,y)
plt.show()
plt.plot(x,y,"or")
plt.grid()
plt.grid(color = 'red',linestyle = '--',linewidth=0.8)
plt.show()
```

```python

import numpy as np
from matplotlib import pyplot as plt
a = np.array([22,87,5,43,56,73,55,67,24,96,56,90,32,27])
plt.hist(a,bins=[0,20,40,60,80,100])
plt.title("Histogram")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
score = [65,74,41,71,112,31]
name = ['Sachin','AOUD','ISCJ','oscj','aoscjn','oisj']
myexplode = [0.2,0.1,0,0,0,0]
plt.pie(score,labels=name,startangle=180,explode=myexplode,autopct='%.1f%%')
plt.title('India Team Score')
plt.show()
weight = [12,23,40,34,27,31]
name = ['Mango','papaya','Gwava','cherries','melon','watermelon']

plt.pie(weight,labels=name,startangle=180,autopct='%.1f%%')
plt.title('Fruit Bazar')
plt.show()
```
