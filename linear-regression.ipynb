```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error,r2_score
```

```python
#create data
np.random.seed(0)
X=np.linspace(1,10,50)
y=3*X+5+np.random.randn(50)*2
```

```python
#step 2 train model
from sklearn.linear_model import LinearRegression
X_reshaped = X.reshape(-1,1)
model = LinearRegression()
model.fit(X_reshaped,y)

y_pred=model.predict(X_reshaped)
```

```python
X
```

```python
#STEP 3: Visulaize
plt.scatter(X,y)
plt.plot(X,y_pred)
plt.xlabel("X")
plt.ylabel("y")
plt.title("Simple Linear Regression")
plt.show()
```

```python
X4=np.linspace(-5,5,100).reshape(-1,1)
y4=np.sin(X4).ravel()
from sklearn.svm import SVR #SUPPORT VECTOR REGRESSION
svr=SVR(kernel='rbf',C=100,gamma=0.1)
svr.fit(X4,y4)
y4_pred=svr.predict(X4)
#VISUALIZE
plt.scatter(X4,y4)
plt.plot(X4,y4_pred)
plt.title("Support Vector Regression")
plt.show()
```

```python
#non linear
X5=np.sort(np.random.rand(80),axis=0)
y5=np.sin(5*X5).ravel();
#train model
from sklearn.tree import DecisionTreeRegressor
tree=DecisionTreeRegressor(max_depth=3)
#RESHAPE X5
X5_reshaped=X5.reshape(-1,1)
tree.fit(X5_reshaped,y5)
y5_pred=tree.predict(X5_reshaped)
```

```python
#STEP 3 VISULAIZE
plt.scatter(X5,y5,label='Actual')

plt.plot(X5,y5_pred,color='red',label='Decision Tree Regression')
plt.title("Decision Tree Regression")
plt.legend()

plt.show()
```
