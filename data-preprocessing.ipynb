# *Outliers***

```python
Q1 = df['minimum nights'].quantile(0.25)
Q3 = df['minimum nights'].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5*IQR
upper = Q3 + 1.5*IQR

df['minimum nights'] = df['minimum nights'].clip(lower, upper)
```

```python
Q1 = df['reviews per month'].quantile(0.25)
Q3 = df['reviews per month'].quantile(0.75)
IQR = Q3 - Q1

df['reviews per month'] = df['reviews per month'].clip(
    Q1 - 1.5*IQR, Q3 + 1.5*IQR
)
```

```python
#boolean masking .keeps the value only within the range and drops the rest of rows out of range.
df = df[(df['Construction year'] >= 1800) & (df['Construction year'] <= 2025)]
```
