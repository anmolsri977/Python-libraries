```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = {
    "hours": [1, 2, 3, 4, 5, 6, 7, 8],
    "marks": [30, 35, 45, 50, 60, 70, 80, 90],
    "pass":  [0, 0, 0, 0, 1, 1, 1, 1]  # 0 = Fail, 1 = Pass
}

df = pd.DataFrame(data)
plt.scatter(df["hours"],df["pass"])
plt.title("pass/fail")
plt.xlabel("hours")
plt.ylabel("pass(1)/fail(0)")
X = df[["marks"]]   # feature
y = df["pass"]      # class label

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.25,
    random_state=42
)
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
model.predict_proba(X_test)




from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_test, y_pred)



from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_pred)


newinput=pd.DataFrame({"marks": [93]})
y_pred = model.predict(newinput)
model.predict_proba(newinput)
print(model.predict_proba(pd.DataFrame({"marks": [30]})))
print(model.predict_proba(pd.DataFrame({"marks": [55]})))
print(model.predict_proba(pd.DataFrame({"marks": [70]})))
print(model.predict_proba(pd.DataFrame({"marks": [90]})))
```
